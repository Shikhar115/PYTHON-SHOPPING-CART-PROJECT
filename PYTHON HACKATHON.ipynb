```python
menu = [['Milkshake',50,80], ['Ice Cream',100,50], ['Chocolate',200,10],[ 'Eclairs  ',300,5],[ 'CupCake  ',150,25]]
```

```python
print('Welcome to the shikhar grocery store !')
print()

# displaying items description


print ('                    Items Available            ')
print()

print('%10s %10s %10s %10s'% ('S.NO','ITEM','QUANTITY','PRICE/ITEM'))
print('%10d %10s %10.2f %10.2f'%(1,menu[0][0],menu[0][1],menu[0][2]))
print('%10d %10s %10.2f %10.2f'%(2,menu[1][0],menu[1][1],menu[1][2]))
print('%10d %10s %10.2f %10.2f'%(3,menu[2][0],menu[2][1],menu[2][2]))
print('%10d %10s %10.2f %10.2f'%(4,menu[3][0],menu[3][1],menu[3][2]))
print('%10d %10s %10.2f %10.2f'%(5,menu[4][0],menu[4][1],menu[4][2]))

print()

# creating list for quantity ordered and cost of each item in order

shopping_quant= []
cost = [] 
    
# taking order input from customer
j=0
while j==0:
    
    order1 = int(input('How many milkshake do you need ? '))
    if order1<=menu[0][1]:
        menu[0][1]-=order1
        shopping_quant.append(order1)
        cost.append(order1*menu[0][2])
        j==1
        break
    elif order1> menu[0][1]:
        shopping_quant.append(0)
        print('insufficient stock, remaining quantity of this item ',menu[0][1])
        a=input('** to proceed, enter p** ---** to make changes, enter c** --- ')
        if a=='c':
            j==0
        elif a=='p':
            j==1
            break

j=0
while j==0:        
    order2 = int(input('How many ice cream do you need ? ')) 
    if order2<=menu[1][1]:
        menu[1][1]-=order2
        shopping_quant.append(order2)
        cost.append(order2*menu[1][2])
        j==1
        break
    elif order2> menu[1][1]:
        shopping_quant.append(0)
        print('insufficient stock, remaining quantity of this item ',menu[1][1])
        a=input(' **to proceed, enter p** --- **to make changes, enter c** --- ')
        if a=='c':
            j==0
        elif a=='p':
            j==1
            break

j=0
while j==0:    
    order3 = int(input('How many chocolate do you need ? ')) 
    if order3<=menu[2][1]:
        menu[2][1]-=order3
        shopping_quant.append(order3)
        cost.append(order3*menu[2][2])
        j==1
        break
    elif order3> menu[2][1]:
        shopping_quant.append(0)
        print('insufficient stock, remaining quantity of this item ',menu[2][1])
        a=input(' **to proceed, enter p** --- **to make changes, enter c** --- ')
        if a=='c':
            j==0
        elif a=='p':
            j==1
            break

j=0
while j==0:    
    order4 = int(input('How many eclairs do you need ? ')) 
    if order4<=menu[3][1]:
        menu[3][1]-=order4
        shopping_quant.append(order4)
        cost.append(order4*menu[3][2])
        j==1
        break
    elif order4> menu[3][1]:
        shopping_quant.append(0)
        print('insufficient stock, remaining quantity of this item ',menu[3][1])
        a=input('** to proceed, enter p** ---** to make changes, enter c **--- ')
        if a=='c':
            j==0
        elif a=='p':
            j==1
            break

j=0
while j==0:    
    order5 = int(input('How many cupcake do you need ? ')) 
    if order5<=menu[4][1]:
        menu[4][1]-=order5
        shopping_quant.append(order5)
        cost.append(order5*menu[4][2])
        j==1
        break
    elif order5> menu[4][1]:
        shopping_quant.append(0)
        print('insufficient stock, remaining quantity of this item ',menu[4][1])
        a=input('** to proceed, enter p** --- **to make changes, enter c **--- ')
        if a=='c':
            j==0
        elif a=='p':
            j==1
            break
    
print()

# taking input of delivery details

a1=input('enter your name : ')
a2=input('enter your address : ')

print()

print('delivery charges of Rs.20 will be levied for distances above 3 km and upto 5km ')
print('delivery charges of Rs.40 will be levied for distances above 5 km')
print()

a3=int(input('your distance from store : '))

# delivery  charge

del_charge=0
if a3<3:
    del_charge=0
elif a3 in range(3,6):
        del_charge = 20
elif a3 >5:
    del_charge = 40
print()
        
# Showing the shopping cart. 

print()
print('Your Shopping Cart:')
print()
print('%10s %10s %10s %10s'% ('S.NO','ITEM','QUANTITY','COST'))
print('%10d %10s %10.2f %10.2f'%(1,menu[0][0],shopping_quant[0],cost[0]))
print('%10d %10s %10.2f %10.2f'%(2,menu[1][0],shopping_quant[1],cost[1]))
print('%10d %10s %10.2f %10.2f'%(3,menu[2][0],shopping_quant[2],cost[2]))
print('%10d %10s %10.2f %10.2f'%(4,menu[3][0],shopping_quant[3],cost[3]))
print('%10d %10s %10.2f %10.2f'%(5,menu[4][0],shopping_quant[4],cost[4]))

print()

# total bill


total_cost= cost[0]+cost[1]+cost[2]+cost[3]+cost[4]
print('total cost : Rs. ',total_cost)

print()

# final amount due considering delivery charge

final_amount = total_cost + del_charge
print('''final amount = total cost : {} + delivery charge : {}
final amount = {}'''.format(total_cost,del_charge,final_amount) )


print()

# greetings

print('name :',a1)
print('address : ',a2)

print()

print('Thanks!')
print('Have a nice day')

```
